```python
import pandas as pd
import os

# Data path
path = '/Users/alexcervino/Desktop/DARKMATTER/channel_h/executions'
os.chdir(path)
csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]
print("CSV files found:", csv_files)

for i in csv_files:
    print("Processing file:", i)
    df = pd.read_csv(i)

    # 2. Define quantiles
    quantiles = [0.05, 0.32, 0.5, 0.68, 0.95]

    # 3. Group by all text columns + mass to preserve information
    #    and calculate quantiles for 'sigma_v' or 'tau'
    if 'annihilation' in i:
        df_final = df.groupby(['src1', 'profile', 'ch', 'mass'])['sigma_v'].quantile(quantiles)
    elif 'decay' in i:
        df_final = df.groupby(['src1', 'profile', 'ch', 'mass'])['tau'].quantile(quantiles)

    df_final = df_final.reset_index()
    df_final = df_final.rename(columns={'level_4': 'quantile'})
    print("Results with corrected naming:")
    print(df_final.head())

    # Merge
    # Extract source name, profile, and type from filename
    file_parts = i.replace('_t100.csv', '').split('_')
    src_name = file_parts[0]

    profile = file_parts[1].lower()
    file_type = 'annihilation' if 'annihilation' in i else 'decay'

    # Build path to matching file
    limits_path = '/Users/alexcervino/Desktop/DARKMATTER/ksp_dsph_material/limits_wsj/'+file_type+'/single'
    limits_file = f'{src_name}_{profile}_{file_type}_t100_wsj_quantiles.csv'
    limits_file_path = os.path.join(limits_path, limits_file)

    # Read and merge with limits file if it exists
    if os.path.exists(limits_file_path):
        print(limits_file_path)
        df_limits = pd.read_csv(limits_file_path)
        df_final = pd.concat([df_final, df_limits], ignore_index=True)
        print(df_final)
    else:
        print(f"Warning: Limits file not found: {limits_file_path}")

    # 6. Save to CSV respecting the requested format
    #    index=True will add the first numeric column (350, 351, etc.)
    df_final.to_csv('/Users/alexcervino/Desktop/DARKMATTER/channel_h/quantiles/'+i+'_quantiles.csv', index=True)

    # Print to verify the first few rows
    print(df_final.head(10))
```
